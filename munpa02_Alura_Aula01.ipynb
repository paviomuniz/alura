```python
b = dict((i, i.replace("-","")) for c,i in enumerate(dados.columns) )
```

```python
dados2 = dados.rename(columns=b)
```

```python
dados2.query('g1 >2 and dose=="D2" and tempo==72')['droga'].value_counts()[:10].plot.bar(title="Drogas com f>2 para g1 em 72h",grid=True,style='mbh')
```

```python
dados2.query('g1 >2 and dose=="D2" and tempo==48')['droga'].value_counts()[:10].plot.bar(title="Drogas com f>2 para g1 em 72h",grid=True,style='mbh')
```

```python
dados2.query('g1 >2 and dose=="D2" and tempo==24')['droga'].value_counts()[:10].plot.bar(title="Drogas com f>2 para g1 em 72h",grid=True,style='mbh')
```
